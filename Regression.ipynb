```python
# import Libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Download and prepare the data

data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
lifesat
```

```python
# Convert table to training examples

X = lifesat[['GDP per capita (USD)']].values
y = lifesat[['Life satisfaction']].values
```

```python
X
```

```python
y
```

```python
# Visualize the data
plt.scatter(X, y)
plt.title('Life satisfaction to GPD per capita')
plt.xlabel('GDP per capita')
plt.ylabel('Life satisfaction')
plt.show()
```

```python
# Model selection 
model = LinearRegression()
```

```python
# Training model
model.fit(X, y)
```

```python
# Make prediction
X_new = [[6766.48]] # South africa GDP per capita
y_hat = model.predict(X_new)
print(y_hat)
```

```python
# Final Visualization
plt.scatter(X, y, label='Training examples')
plt.scatter(X_new, y_hat, color='r', marker='x', label='Predicted value')
plt.legend()
plt.xlabel('GDP per capita')
plt.ylabel('Life satisfaction')
plt.show()
```
